```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

def draw_node(ax, xy, text):
    """画一个圆角矩形节点"""
    box = FancyBboxPatch(
        (xy[0]-0.5, xy[1]-0.25), 1, 0.5,
        boxstyle="round,pad=0.1",
        edgecolor="black", facecolor="lightblue"
    )
    ax.add_patch(box)
    ax.text(xy[0], xy[1], text, ha="center", va="center", fontsize=10)

def draw_arrow(ax, start, end):
    """画箭头"""
    ax.annotate("",
        xy=end, xycoords="data",
        xytext=start, textcoords="data",
        arrowprops=dict(arrowstyle="->", lw=1.2))

# 创建画布
fig, ax = plt.subplots(figsize=(10, 5))

# --- Training Phase (Offline) ---
draw_node(ax, (0, 2), "Episodes\n(Trace Data)")
draw_node(ax, (2.5, 2), "OPT Oracle\n(Labels)")
draw_node(ax, (5, 2), "Train ML Model")

draw_arrow(ax, (0.5, 2), (2, 2))
draw_arrow(ax, (3, 2), (4.5, 2))

ax.text(2.5, 2.6, "Training Phase (Offline)", ha="center", fontsize=12, fontweight="bold")

# --- Online Phase (Admission) ---
draw_node(ax, (0, 0), "Request Miss")
draw_node(ax, (2.5, 0), "Feature\nExtraction")
draw_node(ax, (5, 0), "ML Admission\nModel")
draw_node(ax, (7.5, 0), "Decision\n(Admit / Reject)")

draw_arrow(ax, (0.5, 0), (2, 0))
draw_arrow(ax, (3, 0), (4.5, 0))
draw_arrow(ax, (5.5, 0), (7, 0))

ax.text(3.75, 0.6, "Online Inference Phase", ha="center", fontsize=12, fontweight="bold")

# 美化
ax.set_xlim(-1, 9)
ax.set_ylim(-1, 3)
ax.axis("off")

plt.tight_layout()

# ⚠️ 一定要先保存，再 show
plt.savefig(r"C:\Users\86136\Desktop\FDU\capstone\ml_admission_loop.pdf", 
            format="pdf", bbox_inches="tight")

plt.show()


```
